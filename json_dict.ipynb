# Dictionnaire et json

Différence fondamentale entre **dictionnaire** et **json**, et la confusion est très courante. Pour faire simple :

- Un *dictionnaire* Python est un type de données vivant dans la mémoire de votre programme Python.

- *JSON* est un format de texte utilisé pour stocker et échanger des données.

Voyons cela en détail.

## 1. Le Dictionnaire Python

C'est une structure de données native en Python, utilisée pour stocker des paires clé-valeur.

**Caractéristiques** :

- **Type de données natif** : Fait partie du langage Python (*dict*).

- **En mémoire** : Il existe dans la RAM pendant l'exécution de votre script.

- **Types de données riches** : Les clés peuvent être de n'importe quel type hashable (chaîne, nombre, tuple), et les valeurs peuvent être **n'importe quel objet Python**.

```python
mon_dict = {
    "cle_string": "valeur",
    42: "une clé numérique",     # Clé de type int
    ("tu", "ple"): "une clé de tuple",   # Clé de type tuple
    "une_liste": [1, 2, 3, {"sous-dict": True}],   # Valeur: liste contenant un dict
    "une_fonction": print,  # On peut même stocker une fonction !
    "None": None  # Le type None
}
```

- **Usage** : Manipulation de données à l'intérieur d'un programme Python.

## 2. JSON (JavaScript Object Notation)

C'est un format de sérialisation, c'est-à-dire une convention d'écriture sous forme de chaîne de caractères, dérivé de la syntaxe des objets JavaScript.

**Caractéristiques** :

- **Format texte** : C'est une chaîne de caractères (*str*).

- **Pour la persistance et l'échange** : Il est conçu pour être sauvegardé dans un fichier ou envoyé via un réseau (ex: API web).

- **Types de données limités** : Seuls quelques types sont autorisés, pour garantir la compatibilité entre différents langages.

-- *String* (Chaîne de caractères)

-- *Number* (Nombre)

-- *Boolean* (true / false)

-- *Array* (Tableau)

-- *Object* (Objet, équivalent à un dictionnaire)

-- *null*

- **Clés obligatoirement entre doubles guillemets** :

```python
{
    "nom": "Alice",
    "age": 30,
    "hobbies": ["lecture", "rando"],
    "actif": true,
    "adresse": null
}
```

**Note**: *true*, *false* et *null* sont en minuscules.

```python
# Tableau Comparatif Direct
""" 
Aspect	:                   Dictionnaire Python	                                          JSON
Nature  :	          Structure de données en mémoire	                          Format de texte (chaîne)
Type Python	:                    dict	                                              str
Clés   :	           N'importe quel type hashable	                              Uniquement des strings
Guillemets : 	       Simple ' ou double " acceptés	                          Obligatoirement doubles "
Types de valeurs : 	Tous objets Python (list, function, custom class, etc.) 	Types limités (string, number, array, object, boolean, null)
Booléens  :           	True, False (avec une majuscule)	                      true, false (en minuscule)
Valeur nulle :                	None	                                                     null
Usage principal	 :    Manipulation interne dans un programme	             Stockage, échange de données entre systèmes

"""
```

## Le Lien entre les deux : Sérialisation et Désérialisation

C'est là que la magie opère. On peut convertir un dictionnaire Python en JSON (sérialisation) et inversement (désérialisation) grâce au module *json* intégré à Python.


### Conversion Dictionnaire → JSON (*json.dumps()*)

```python
import json

# Un dictionnaire Python
personne_dict = {
    "nom": "Bob",
    "age": 25,
    "enfants": ["Anna", "Tom"],
    "marie": True,
    "voiture": None
}

# Sérialisation en chaîne JSON
json_string = json.dumps(personne_dict, indent=4)    # indent pour la lisibilité pour les humains
print(json_string)
print(type(json_string))    # <class 'str'>
```

Notez que *True* est devenu *true* et *None* est devenu *null*.

### Conversion JSON → Dictionnaire (*json.loads()*)

```python
# Une chaîne JSON (qu'on pourrait lire dans un fichier ou recevoir d'une API)
json_data = '{"nom": "Alice", "langage_prefere": "Python", "actif": true, "projets": null}'

# Désérialisation en dictionnaire Python
python_dict = json.loads(json_data)
print(python_dict)   # {'nom': 'Alice', 'langage_prefere': 'Python', 'actif': True, 'projets': None}
print(type(python_dict))    # <class 'dict'>
print(python_dict["nom"])    # Alice
```

## Cas Pratiques et Pièges Courants

1 **Clés non-strings** : Un dictionnaire Python avec une clé *int* sera converti en une clé *string* en JSON.

```python
mon_dict = {1: "un"}
print(mon_dict)  # {1: 'un'}  La clé est un entier dans le dict Python

print()
json_str = json.dumps(mon_dict) 
print(json_str)  # {"1": "un"}  La clé 1 a été convertie en string "1" dans le JSON, 
                         # car JSON n'autorise que les clés de type string.
print()
nouveau_dict = json.loads(json_str)  
print(nouveau_dict)   # {"1": "un"} -> la clé est maintenant une string !
print(type(nouveau_dict))  # <class 'dict'>
```

2 **Objets complexes** : 

Vous ne pouvez pas sérialiser directement un objet de classe personnalisée, une fonction ou un *set*.

```python
import json

class User:
    def __init__(self, name):
        self.name = name
    

user = User("Charlie")

# json_user = json.dumps(user)  # Erreur : Object of type User is not JSON serializable



# Solution simple : sérialiser le dict interne de l'objet qui contient ses attributs et leurs valeurs 
json_user = json.dumps(user.__dict__)  # Sérialisation du dict interne de l'objet
print(json_user)      # {"name": "Charlie"} , ceci est une chaîne JSON représentant les attributs de l'objet User

```

Autre alternative : 

Il faut fournir une fonction personnalisée au paramètre *default* de *dumps* pour convertir l'objet en un dictionnaire sérialisable.

# Conclusion

Pour résumer :

- Vous utilisez un **dictionnaire** pour travailler avec les données à l'**intérieur** de votre code Python.

- Vous utilisez **JSON** pour **sauvegarder** ces données dans un fichier ou les **envoyer** à un autre programme (comme un frontend JavaScript ou une API).

- Le module *json* est le pont qui permet de faire la navette entre ces deux mondes.

# Un cas cconcret simple : météo

Qu'est-ce qu'une *API* météo ?

Imaginez : l'*API* météo est comme un robot qui connaît la météo du monde entier.
 
Vous lui posez une question ("Quel temps fait-il à Paris ?") et il vous répond en *JSON* (un texte structuré)

```python
"""  
Vous (votre programme) --- question ---> API météo (le robot)
                                          ↓
Vous (votre programme) <--- réponse JSON --- API météo
"""
```

### Exemple simple

```python
import urllib.request
import json

# 1. L'URL de l'API (la question qu'on pose au robot)
url = "https://api.open-meteo.com/v1/forecast?latitude=48.8566&longitude=2.3522&current_weather=true"

# 2. On pose la question au robot
try:
    reponse = urllib.request.urlopen(url)
    print("Requête réussie, code HTTP :", reponse.status)
except Exception as e:
    print("Erreur lors de la requête à l'API météo :", e)
    exit(1)    # On arrête le programme en cas d'erreur de requête



# 3. On lit la réponse (c'est du texte JSON)
donnees_brutes = reponse.read().decode('utf-8')  # decode pour convertir les bytes en string car read() retourne des bytes

# 4. On transforme le texte JSON en dictionnaire Python
meteo = json.loads(donnees_brutes )

# 5. On affiche ce qu'on a trouvé
print(meteo)
```

Résultat est un gros dictionnaire

### Extraire seulement ce qui nous intéresse

```python
# On extrait juste la météo actuelle
actuel = meteo['current_weather']

print(f"🌡️  Température : {actuel['temperature']}°C")
print(f"💨 Vent : {actuel['windspeed']} km/h")
print(f"🕐 Heure : {actuel['time']}")
```

### Pourquoi ces nombres bizarres (48.8566, 2.3522) ?

L'*API* météo ne comprend pas les noms de villes ("Paris"), seulement les coordonnées GPS (latitude, longitude).

- **Paris** : latitude 48.8566, longitude 2.3522

- **Londres** : latitude 51.5074, longitude -0.1278

- **New York** : latitude 40.7128, longitude -74.0060

### Le programme complet (enfin !)

Assemblons tout :

```python
import urllib.request
import json

def trouve_coordonnees(nom_ville):
    """Étape 1 : Transformer un nom de ville en coordonnées"""
    ville_encodee = nom_ville.replace(" ", "%20")
    url_geo = f"https://geocoding-api.open-meteo.com/v1/search?name={ville_encodee}&count=1&format=json"
    
    reponse = urllib.request.urlopen(url_geo)
    donnees = reponse.read().decode('utf-8')
    resultat = json.loads(donnees)
    
    if resultat.get('results'):
        premiere_ville = resultat['results'][0]
        return {
            'nom': premiere_ville['name'],
            'pays': premiere_ville['country'],
            'latitude': premiere_ville['latitude'],
            'longitude': premiere_ville['longitude']
        }
    return None


def get_meteo(latitude, longitude):
    """Étape 2 : Récupérer la météo pour des coordonnées"""
    url_meteo = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"
    
    reponse = urllib.request.urlopen(url_meteo)
    donnees = reponse.read().decode('utf-8')
    return json.loads(donnees)


#===================  Programme principal ==================================
print("🌤️  MÉTEO SIMPLE")
print("=" * 30)

nom_ville = input("Entrez un nom de ville : ")

# Étape 1 : Trouver les coordonnées
print(f"🔍 Recherche de {nom_ville}...")
coordonnees = trouve_coordonnees(nom_ville)

if coordonnees:
    print(f"✅ Ville trouvée : {coordonnees['nom']}, {coordonnees['pays']}")
    
    # Étape 2 : Récupérer la météo
    print(f"🌤️  Récupération de la météo...")
    meteo = get_meteo(coordonnees['latitude'], coordonnees['longitude'])
    
    # Étape 3 : Afficher les résultats
    actuel = meteo['current_weather']
    print(f"\n   Dernière mise à jour : {actuel['time']}")
    print(f"🌡️  Température : {actuel['temperature']}°C")
    print(f"💨 Vent : {actuel['windspeed']} km/h")
    
    # Petit bonus : interpréter le code météo
    codes_meteo = {
        0: "☀️ Ciel clair",
        1: "🌤️ Principalement clair", 
        2: "⛅ Partiellement nuageux",
        3: "☁️ Couvert",
        45: "🌫️ Brouillard",
        51: "🌧️ Bruine",
        61: "🌧️ Pluie",
        71: "❄️ Neige",
        95: "⛈️ Orage"
    }
    description = codes_meteo.get(actuel['weathercode'], "❓ Inconnu")
    print(f"☁️  Conditions : {description}")
else:
    print(f"❌ Désolé, je n'ai pas trouvé '{nom_ville}'")
```

### Sauvegarder la météo dans un fichier

Ajoutons une fonction pour sauvegarder :

```python
def sauvegarde_meteo(nom_ville, meteo):
    """Sauvegarde la météo dans un fichier JSON"""
    
    # On crée un dictionnaire avec toutes les infos
    donnees_sauvegardees = {
        "ville": nom_ville,
        "date": meteo['current_weather']['time'],
        "temperature": meteo['current_weather']['temperature'],
        "vent": meteo['current_weather']['windspeed'],
        "code_meteo": meteo['current_weather']['weathercode']
    }
    
    # On sauvegarde dans un fichier
    nom_fichier = f"meteo_{nom_ville.lower().replace(' ', '_')}.json"
    
    with open(nom_fichier, 'w', encoding='utf-8') as fichier:
        json.dump(donnees_sauvegardees, fichier, indent=2, ensure_ascii=False)
    
    print(f"💾 Météo sauvegardée dans {nom_fichier}")
    return nom_fichier


```

### Lire un fichier JSON sauvegardé

```python
def lit_meteo_sauvegardee(nom_fichier):
    """Lit un fichier JSON de météo sauvegardé"""
    try:
        with open(nom_fichier, 'r', encoding='utf-8') as fichier:
            donnees = json.load(fichier)
        print(f"\n📂 Données chargées depuis {nom_fichier}")
        print(f"Ville : {donnees['ville']}")
        print(f"Température : {donnees['temperature']}°C")
        print(f"Date : {donnees['date']}")
        return donnees
    except FileNotFoundError:
        print(f"❌ Fichier {nom_fichier} non trouvé")
        return None

# Testez après avoir sauvegardé :
# lit_meteo_sauvegardee("meteo_nancy.json")
```

```python
#===================  Programme principal ==================================
print("🌤️  MÉTEO SIMPLE")
print("=" * 30)

nom_ville = input("Entrez un nom de ville : ")

# Étape 1 : Trouver les coordonnées
print(f"🔍 Recherche de {nom_ville}...")
coordonnees = trouve_coordonnees(nom_ville)

if coordonnees:
    print(f"✅ Ville trouvée : {coordonnees['nom']}, {coordonnees['pays']}")
    
    # Étape 2 : Récupérer la météo
    print(f"🌤️  Récupération de la météo...")
    meteo = get_meteo(coordonnees['latitude'], coordonnees['longitude'])
    
    # Étape 3 : Afficher les résultats
    actuel = meteo['current_weather']
    print(f"\n   Dernière mise à jour : {actuel['time']}")
    print(f"🌡️  Température : {actuel['temperature']}°C")
    if actuel['temperature'] < 0:
        print("❄️ Il gèle !")


    print(f"💨 Vent : {actuel['windspeed']} km/h")
    print(f"🧭 Direction du vent : {actuel['winddirection']}°")
    
    # Petit bonus : interpréter le code météo
    codes_meteo = {
        0: "☀️ Ciel clair",
        1: "🌤️ Principalement clair", 
        2: "⛅ Partiellement nuageux",
        3: "☁️ Couvert",
        45: "🌫️ Brouillard",
        51: "🌧️ Bruine",
        61: "🌧️ Pluie",
        71: "❄️ Neige",
        95: "⛈️ Orage"
    }
    description = codes_meteo.get(actuel['weathercode'], "❓ Inconnu")
    print(f"☁️  Conditions : {description}")
else:
    print(f"❌ Désolé, je n'ai pas trouvé '{nom_ville}'")


sauvegarder = input("\nSauvegarder cette météo ? (o/n) : ")
if sauvegarder.lower() == 'o':
    sauvegarde_meteo(coordonnees['nom'], meteo)


```

## Points clés à retenir de ce cas pratique :
