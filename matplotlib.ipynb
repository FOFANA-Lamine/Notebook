## 3. Création de Figures et Sous-graphiques

Figure de base

```python
plt.figure(figsize=(8, 4))  # Largeur, hauteur en pouces
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Graphique de base")
plt.xlabel("Axe X")
plt.ylabel("Axe Y")
plt.grid(True)
plt.show()
```

Sous-graphiques (subplots)

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
axes[0, 0].plot(np.random.rand(10))
axes[0, 0].set_title("Graphique 1")
axes[0, 1].scatter(np.random.rand(10), np.random.rand(10))
axes[0, 1].set_title("Graphique 2")
axes[1, 0].bar(['A', 'B', 'C'], [3, 7, 2])
axes[1, 0].set_title("Graphique 3")
axes[1, 1].hist(np.random.randn(1000), bins=20)
axes[1, 1].set_title("Graphique 4")
plt.tight_layout()  # Évite le chevauchement
plt.show()
```

## 4. Types de Graphiques

Ligne (plot)

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', color='red', linestyle='-', linewidth=1)
plt.legend()
plt.show()
```

Nuage de points (scatter)

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar()  # Affiche la barre de couleur
plt.show()
```

Histogramme

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title("Histogramme avec densité")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()
```

Barres

```python
categories = ['Catégorie A', 'Catégorie B', 'Catégorie C']
values = [15, 25, 30]

plt.bar(categories, values, color=['red', 'green', 'blue'])
# plt.barh(categories, values)  # Version horizontale
plt.show()
```

Camembert (pie chart)

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 20, 15, 20]
explode = (0.1, 0, 0, 0)  # Mise en évidence de Python

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Assure que le cercle est parfait
plt.title("Parts de marché des langages")
plt.show()
```

Boîtes à moustaches (boxplot)

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True,
           labels=['Groupe 1', 'Groupe 2', 'Groupe 3'])
plt.title("Boxplot multiple")
plt.show()
```

Violon (violin plot)

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title("Violin plot")
plt.show()
```

## 6. Graphiques 3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)  # Crée une grille 2D
Z = np.sin(np.sqrt(X**2 + Y**2))   

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Surface 3D")
ax.set_xlabel("Axe X")
ax.set_ylabel("Axe Y")
ax.set_zlabel("Axe Z")
plt.show()
```

## 7. Visualisation de Données Multidimensionnelles

Matrice de corrélation

```python
import pandas as pd

# np.random.randn(100, 4) = 100 lignes, 4 colonnes de données aleatoires gaussiennes
df = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])  
corr = df.corr() 
plt.imshow(corr, cmap='coolwarm', interpolation='none')  
plt.colorbar()  
plt.xticks(range(len(corr)), corr.columns)   # xticks() / yticks() : Définissent les étiquettes des axes
plt.yticks(range(len(corr)), corr.columns)
plt.title("Matrice de corrélation")
plt.show()
```

## Explication :

*.corr()* : Méthode qui calcule les coefficients de corrélation entre toutes les paires de colonnes
Retourne une matrice carrée où chaque cellule représente la corrélation entre deux variables
Coefficient de corrélation : Valeur entre -1 et 1

1 : Corrélation positive parfaite

-1 : Corrélation négative parfaite

0 : Aucune corrélation linéaire



*imshow()* : Fonction pour afficher une matrice comme une image

*cmap='coolwarm'* : Palette de couleurs ("cool" pour valeurs négatives, "warm" pour positives)

*interpolation='none'* : Désactive le lissage pour garder les pixels nets

*plt.colorbar()* : Ajoute une échelle de couleurs à côté du graphique.

Permet de faire la correspondance entre les couleurs et les valeurs numériques

-------------

Heatmap

```python
import seaborn as sns  

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Heatmap de corrélation")
plt.show()
```

## 8. Animation avec Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import warnings
warnings.filterwarnings('ignore')  # Masquer les avertissements


fig, ax = plt.subplots(figsize=(8, 4))
xdata, ydata = [], []   # Listes vides pour stocker les points au fur et à mesure
ln, = ax.plot([], [], 'ro', markersize=3)  # Crée un tracé vide

# Fonction d'initialisation
def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('x')
    ax.set_ylabel('sin(x)')
    ax.set_title('Animation de la fonction sinus')
    return ln,

# Fonction de mise à jour
def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    
    # Optionnel : Mettre à jour les limites si nécessaire
    if len(xdata) > 0:
        ax.set_xlim(min(xdata) - 0.1, max(xdata) + 0.1)
    
    return ln,

# Créer l'animation
ani = FuncAnimation(fig, update, 
                    frames=np.linspace(0, 2*np.pi, 64),  # Réduit le nombre de frames
                    init_func=init,    
                    blit=True,   # Optimisation (redessine seulement les parties modifiées)
                    interval=50,  # Intervalle en ms
                    repeat=False)


# ani.to_jshtml() : Convertit l'animation en HTML avec JavaScript
HTML(ani.to_jshtml())  # Affiche le contenu HTML dans Jupyter

# Alternative : Sauvegarder en GIF (si pillow est installé)
# ani.save('sin_animation.gif', writer='pillow', fps=20)
```

## Explication :

- Fonction init() - Initialisation

*init()* : Fonction appelée au début de l'animation

*set_xlim(0, 2*np.pi)* : Limites de l'axe X (0 à 2π)

*set_ylim(-1, 1)* : Limites de l'axe Y (-1 à 1)

*grid(True, alpha=0.3)* : Active la grille avec transparence

*set_xlabel(), set_ylabel(), set_title()* : Labels et titre

*return ln,* : Retourne l'objet à animer (note: la virgule fait un tuple)


- Fonction update() - Mise à jour

*update(frame)* : Fonction appelée à chaque frame

*frame* : Valeur courante fournie par frames (ici, de 0 à 2π)

*xdata.append(frame)* : Ajoute la coordonnée X

*ydata.append(np.sin(frame))* : Calcule et ajoute sin(x)

*ln.set_data(xdata, ydata)* : Met à jour les données du tracé

- Mise à jour dynamique des limites :

*ax.set_xlim(min(xdata) - 0.1, max(xdata) + 0.1)*

L'axe X s'ajuste automatiquement (+/- 0.1 de marge)

*return ln,* : Retourne l'objet modifié



- Déroulement de l'animation :

Frame 0 : *init()* est appelée → Graphique vide avec grille

Frame 1 : *update(0)* → Point (0, sin(0)=0)

Frame 2 : *update(0.1)* → Points (0,0) et (0.1, 0.0998)

...

Frame 64 : *update(2π)* → Tous les points de la sinusoïde

## 11. Astuces Avancées

Twin Axes (deux axes Y)

```python
fig, ax1 = plt.subplots()

x = np.arange(0, 10, 0.1)
y1 = np.exp(x)
y2 = np.sin(x)

color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('exp(x)', color=color)
ax1.plot(x, y1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Crée un deuxième axe partageant le même axe x

color = 'tab:blue'
ax2.set_ylabel('sin(x)', color=color)
ax2.plot(x, y2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Deux axes y")
plt.show()
```

Grilles personnalisées

```python
plt.figure(figsize=(8, 6))
ax = plt.gca()

ax.set_facecolor('#f0f0f0')  # Couleur de fond
ax.grid(which='major', linestyle='-', linewidth='0.5', color='white')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='white')

# Détails mineurs
ax.minorticks_on()


plt.plot(np.random.randn(100).cumsum(), color='blue')  
plt.title("Graphique avec grille personnalisée")
plt.show()
```

Insets (graphiques dans des graphiques)

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Graphique principal
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), 'b-')

# Inset (Zoom) de graphique secondaire sur le graphique principal
inset_ax = fig.add_axes([0.2, 0.6, 0.25, 0.25])  # [x, y, width, height]
inset_ax.plot(x, np.cos(x), 'r-')
inset_ax.set_title('Zoom cos(x)')

plt.show()
```

## 12. Optimisation des Performances

Pour les grandes quantités de données :


```python

from matplotlib.collections import LineCollection   # Collection de lignes

# Méthode standard (lente pour beaucoup de points (>10 000))  
# plt.plot(x, y)

# Méthode optimisée
points = np.array([x, y]).T.reshape(-1, 1, 2)  # .reshape(-1, 1, 2) : Réorganise les dimensions
segments = np.concatenate([points[:-1], points[1:]], axis=1)

lc = LineCollection(segments, cmap='viridis')
lc.set_array(np.linspace(0, 1, len(x)))
plt.gca().add_collection(lc)
plt.xlim(x.min(), x.max())
plt.ylim(y.min(), y.max())
plt.colorbar(lc)
plt.show()
```

## 13. Thèmes et Styles Personnalisés
