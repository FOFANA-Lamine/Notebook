## 1. Introduction à Pandas

Pandas fournit des structures de données et des outils d'analyse de données hautes performances. Ses deux structures principales sont les Series (1D) et les DataFrames (2D).

```python
import pandas as pd
import numpy as np
```

## 2. Structures de données fondamentales

Series (séries de données)

Series pandas - vecteur unidimensionnel

Une Series est comme une colonne unique d'un DataFrame.
C'est l'objet fondamental pour les opérations vectorisées pandas.

```python
print("1. SERIES SIMPLE AVEC INDEX PAR DÉFAUT :")
s1 = pd.Series([10, 25, 30, 42, 55])
print(s1)

print("\n2. SERIES AVEC INDEX PERSONNALISÉ :")
s2 = pd.Series([100, 150, 200, 250], 
               index=['Paris', 'Lyon', 'Marseille', 'Toulouse'],
               name='ventes_2024')
print(s2)

print("\n3. SERIES AVEC DONNÉES MANQUANTES (NA) :")
s3 = pd.Series([12.5, np.nan, 18.3, None, 22.7, np.nan])
print(s3)
```

```python
print("\n4. SERIES À PARTIR D'UN DICTIONNAIRE :")
populations = {'Paris': 2161000, 'Lyon': 518000, 'Marseille': 861000}
s4 = pd.Series(populations, name='population')
print(s4)

print("\n5. SERIES AVEC TYPES SPÉCIFIQUES :")
s5 = pd.Series([1, 2, 3, 4], dtype='float32')
print(f"Type: {s5.dtype}")
print(s5)

print("\n6. OPERATIONS VECTORISEES SUR SERIES :")
temperatures = pd.Series([22.5, 24.0, 19.5, 21.0, 23.5])
print("Températures originales (°C):")
print(temperatures)
print("\nTempératures en Fahrenheit (vectorisé):")
print((temperatures * 9/5) + 32)
```

```python
print("\n7. ACCÈS ET SLICING :")
print("Première valeur :", s2.iloc[0])
print("Valeur pour 'Lyon' :", s2.loc['Lyon'])
print("Deux premières villes :")
print(s2.head(2))

print("\n8. FILTRAGE AVEC CONDITIONS :")
print("Villes avec plus de 150 ventes :")
print(s2[s2 > 150])
```

## 6. Manipulation des données

Ajout/suppression

```python
df['nouvelle_col'] = df['colonne'] * 2    # Ajout colonne
df.drop('colonne', axis=1)          # Suppression colonne
df.drop([0, 1], axis=0)             # Suppression lignes
```


Traitement des valeurs manquantes

```python
# 1. Remplacer les NA par 0

df_remplacer_0 = df.fillna(0)
print(df_remplacer_0)

# 2. Remplacer les NA par la valeur précédente (forward fill)
# Propagation vers l'avant (ffill)
df_ffill = df.ffill()


# df_ffill = df.fillna(method='ffill')
print(df_ffill)

# 3. Suppression des lignes avec NA

df_sans_na = df.dropna()
print(f"Lignes restantes : {len(df_sans_na)} (sur {len(df)})")
print(df_sans_na)

# 4. Interpolation linéaire
df_interpole = df.copy()
# Interpolation uniquement sur les colonnes numériques
df_interpole['Quantité'] = df['Quantité'].interpolate()
df_interpole['Prix_Unitaire'] = df['Prix_Unitaire'].interpolate()
df_interpole['Réduction'] = df['Réduction'].interpolate()
print(df_interpole)

# 5. Autres méthodes de fillna
# Remplacer par des valeurs spécifiques par colonne

df_remplacer_specifique = df.fillna({
    'Quantité': df['Quantité'].mean(),
    'Prix_Unitaire': df['Prix_Unitaire'].median(),
    'Ville': 'INCONNU',
    'Client_Fidèle': False,
    'Réduction': 0
})
print(df_remplacer_specifique)

# 6. Propagation vers l'arrière (backward fill)

df_bfill = df.fillna(method='bfill')
print(df_bfill)

# 7. Suppression des colonnes avec trop de NA
# Par exemple suppression des colonnes avec plus de 50% de NA

df_colonnes_propres = df.dropna(axis=1, thresh=len(df)*0.5)
print(f"Colonnes conservées : {df_colonnes_propres.columns.tolist()}")
```

```python
# Création du DataFrame du debut sans valeurs manquantes

df = pd.DataFrame({
    'Date': pd.date_range('2026-01-01', periods=12, freq='D'),
    'Produit': ['Ordinateur', 'Téléphone', 'Tablette', 'Ordinateur', 
                'Téléphone', 'Écouteurs', 'Clavier', 'Souris',
                'Ordinateur', 'Tablette', 'Téléphone', 'Écouteurs'],
    'Catégorie': pd.Categorical(['Électronique', 'Électronique', 'Électronique', 
                                 'Électronique', 'Électronique', 'Accessoire', 
                                 'Accessoire', 'Accessoire',
                                 'Électronique', 'Électronique', 'Électronique', 'Accessoire']),
    'Quantité': [2, 5, 3, 1, 4, 8, 6, 5, 3, 2, 7, 4],
    'Prix_Unitaire': [899.99, 699.50, 349.99, 899.99, 699.50, 79.99, 49.99, 29.99,
                      899.99, 349.99, 699.50, 79.99],
    'Ville': ['Paris', 'Lyon', 'Paris', 'Marseille', 'Paris', 'Lyon',
              'Paris', 'Marseille', 'Lyon', 'Paris', 'Marseille', 'Paris'],
    'Client_Fidèle': [True, False, True, False, True, True, False, True, False, True, True, False]
})
```

```python

# Ajout d'une colonne Total
df['Total'] = df['Quantité'] * df['Prix_Unitaire']


# 1. Groupement par catégorie
grouped_categorie = df.groupby('Catégorie')
print("Moyenne des quantités par catégorie :")
print(grouped_categorie['Quantité'].mean())
print("\nMoyenne des prix par catégorie :")
print(grouped_categorie['Prix_Unitaire'].mean())
```

```python
# 2. Agrégations multiples (par ville)
grouped_ville = df.groupby('Ville')
agg_result = grouped_ville.agg({
    'Quantité': ['sum', 'mean', 'min', 'max'],
    'Total': ['sum', 'mean'],
    'Client_Fidèle': 'sum'  # Nombre de clients fidèles
})
print(agg_result)
```

```python
# 3. Groupement par plusieurs colonnes (par Catégorie et Client Fidèle)
grouped_multi = df.groupby(['Catégorie', 'Client_Fidèle'])
print("Total des ventes par catégorie et type de client :")
print(grouped_multi['Total'].sum().unstack())
```

```python
# 4. Groupement temporel par mois (ventes par mois)
df_temporel = df.copy()
df_temporel['Date'] = pd.to_datetime(df_temporel['Date'])
colonnes_numeriques = df_temporel.select_dtypes(include=[np.number]).columns
grouped_mois = df_temporel.groupby(pd.Grouper(key='Date', freq='ME'))[colonnes_numeriques].sum()
print("Ventes totales par mois :")
print(grouped_mois)



# Alternative 
grouped_mois_alt = df_temporel.groupby(pd.Grouper(key='Date', freq='ME'))[['Quantité', 'Total']].sum()
print("\nAlternative - Ventes par mois (colonnes spécifiques) :")
print(grouped_mois_alt)
```

```python
# 5. Groupement par semaine
grouped_semaine = df_temporel.groupby(pd.Grouper(key='Date', freq='W')).agg({
    'Quantité': 'sum',
    'Total': 'sum',
    'Produit': lambda x: list(x)  # Liste des produits vendus
})
print("Ventes par semaine :")
print(grouped_semaine)
```

```python
# 6. Groupement avec calcul personnalisé (statistiques détaillées par produit)
grouped_produit = df.groupby('Produit').agg({
    'Quantité': ['count', 'sum', 'mean', 'std'],
    'Prix_Unitaire': ['min', 'max', 'mean'],
    'Total': 'sum',
    'Ville': lambda x: ', '.join(sorted(set(x)))  # Villes distinctes
})
print(grouped_produit)
```

```python
# 7. Groupement par ville et produit (top des produits par ville)
grouped_ville_produit = df.groupby(['Ville', 'Produit']).agg({
    'Quantité': 'sum',
    'Total': 'sum'
}).sort_values(by='Total', ascending=False)
print(grouped_ville_produit.head(10))
```

Tableaux croisés

```python
# 1. Crosstab simple (Nombre de ventes par Ville et Catégorie)
cross_simple = pd.crosstab(df['Ville'], df['Catégorie'])
print(cross_simple)

print("-" * 40)

# 2. Crosstab avec totaux
cross_margins = pd.crosstab(df['Ville'], df['Catégorie'], margins=True)
print(cross_margins)
```

```python
# 3. Crosstab avec plusieurs colonnes (Ville × (Catégorie + Client_Fidèle))
cross_multi = pd.crosstab(df['Ville'], [df['Catégorie'], df['Client_Fidèle']])
print(cross_multi)

print("-" * 40)

# 4. Pivot table avec moyenne (Prix moyen par Catégorie et Client_Fidèle)
pivot_mean = pd.pivot_table(df, 
                           values='Prix_Unitaire', 
                           index='Catégorie', 
                           columns='Client_Fidèle', 
                           aggfunc=np.mean)
print(pivot_mean)
```

```python
# 5. Pivot table avec multiple aggrégations (Statistiques par Catégorie et Ville)
pivot_multi = pd.pivot_table(df,
                            values=['Quantité', 'Prix_Unitaire'],
                            index='Catégorie',
                            columns='Ville',
                            aggfunc={'Quantité': np.sum, 'Prix_Unitaire': np.mean})
print(pivot_multi)

print("-" * 40)

# 6. Pivot table avec remplissage des NaN
pivot_fill = pd.pivot_table(df,
                           values='Prix_Unitaire',
                           index='Catégorie',
                           columns='Client_Fidèle',
                           aggfunc=np.mean,
                           fill_value=0)
print(pivot_fill)
```

```python
# 7. Pivot table avec totaux
pivot_margins = pd.pivot_table(df,
                              values='Quantité',
                              index='Catégorie',
                              columns='Ville',
                              aggfunc=np.sum,
                              margins=True)
print(pivot_margins)
```

```python


# Création de deux DataFrames pour les jointures
df_clients = pd.DataFrame({
    'Client_ID': [1, 2, 3, 4, 5],
    'Nom': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Ville': ['Paris', 'Lyon', 'Paris', 'Marseille', 'Lyon'],
    'Segment': ['Premium', 'Standard', 'Premium', 'Standard', 'Premium']
})


df_commandes = pd.DataFrame({
    'Commande_ID': [101, 102, 103, 104, 105, 106],
    'Client_ID': [1, 2, 1, 3, 6, 4],  # Note: Client_ID 6 n'existe pas dans df_clients
    'Produit': ['Ordinateur', 'Téléphone', 'Tablette', 'Clavier', 'Souris', 'Écouteurs'],
    'Montant': [899.99, 699.50, 349.99, 49.99, 29.99, 79.99],
    'Date': pd.date_range('2026-01-01', periods=6, freq='D')
})

print("DATAFRAME CLIENTS :")
print(df_clients)
print("\n" + "="*50)
print("DATAFRAME COMMANDES :")
print(df_commandes)


```

```python
# 1. INNER JOIN (intersection)
print("1. INNER JOIN - Commandes avec informations clients :")
inner_join = pd.merge(df_commandes, df_clients, on='Client_ID', how='inner')
print(inner_join)

# 2. LEFT JOIN (toutes les commandes)
print("\n2. LEFT JOIN - Toutes les commandes avec infos clients si disponibles :")
left_join = pd.merge(df_commandes, df_clients, on='Client_ID', how='left')
print(left_join)
```

```python
# 3. RIGHT JOIN (tous les clients)
print("\n3. RIGHT JOIN - Tous les clients avec leurs commandes :")
right_join = pd.merge(df_commandes, df_clients, on='Client_ID', how='right')
print(right_join)
```

```python
# 4. FULL OUTER JOIN (union)
print("\n4. FULL OUTER JOIN - Tout combiner :")
outer_join = pd.merge(df_commandes, df_clients, on='Client_ID', how='outer')
print(outer_join)
```

```python

# 1. Conversion en datetime
print("1. CONVERSION DE STRING EN DATETIME :")
df['date_datetime'] = pd.to_datetime(df['date_str'], format='%d/%m/%Y')
print("Types des colonnes de date :")
print(df[['date_str', 'date_datetime']].dtypes)
print(f"\nDate min : {df['date_datetime'].min()}")
print(f"Date max : {df['date_datetime'].max()}")
```

```python
# 2. Extraction de composants temporels
print("\n2. EXTRACTION DE COMPOSANTS TEMPORELS :")
df['annee'] = df['date_datetime'].dt.year
df['mois'] = df['date_datetime'].dt.month
df['nom_mois'] = df['date_datetime'].dt.month_name(locale='fr_FR')
df['jour'] = df['date_datetime'].dt.day
df['jour_semaine'] = df['date_datetime'].dt.day_name(locale='fr_FR')
df['semaine_annee'] = df['date_datetime'].dt.isocalendar().week
df['trimestre'] = df['date_datetime'].dt.quarter

print("Exemple d'extraction :")
print(df[['date_str', 'annee', 'nom_mois', 'jour_semaine', 'trimestre']].head())
```

```python
# 4. Rééchantillonnage temporel (resampling)
print("\n4. RÉÉCHANTILLONNAGE TEMPOREL :")

# Préparation : index temporel et calcul du chiffre d'affaires
df_resample = df.copy()
df_resample = df_resample.set_index('date_datetime')
df_resample['CA'] = df_resample['quantite'] * df_resample['prix']

print("a) Chiffre d'affaires mensuel :")
ca_mensuel = df_resample['CA'].resample('ME').sum()
print(ca_mensuel.head())

print("\nb) Nombre de ventes hebdomadaire :")
ventes_hebdo = df_resample['quantite'].resample('W').sum()
print(ventes_hebdo.head())

print("\nc) Prix moyen quotidien :")
prix_moyen_jour = df_resample['prix'].resample('D').mean()
print(prix_moyen_jour.head())

print("\nd) Statistiques mensuelles complètes :")
stats_mensuelles = df_resample.resample('ME').agg({
    'quantite': 'sum',
    'prix': 'mean',
    'CA': ['sum', 'mean', 'count']
})
print(stats_mensuelles.head())

```

```python
# 5. Filtrage temporel
print("\n5. FILTRAGE TEMPOREL :")

# S'assurer que l'index est trié
df_resample = df_resample.sort_index()

print("a) Ventes du premier trimestre 2026 :")
ventes_T1 = df_resample.loc['2026-01':'2026-03']
print(f"Nombre de ventes T1 : {len(ventes_T1)}")

print("\nb) Ventes du mois de juin 2026 :")
ventes_juin = df_resample.loc['2026-06']
print(f"Nombre de ventes juin : {len(ventes_juin)}")

print("\nc) Ventes des 30 derniers jours (par rapport à la date max) :")
date_max = df_resample.index.max()
date_30j = date_max - pd.Timedelta(days=30)
ventes_30j = df_resample.loc[date_30j:date_max]
print(f"Nombre de ventes 30 derniers jours : {len(ventes_30j)}")
```

   - Utile pour lisser les fluctuations, détecter tendances


2. SOMME CUMULATIVE (*expanding*) :
   - Calcule une statistique qui s'accumule depuis le début
   - Ex: somme cumulative → 10, 25, 37, 55... (10, 10+15, 10+15+12, etc.)
   - Utile pour suivre l'évolution cumulative

Transformation par groupe

```python

df.groupby('groupe')['data'].transform('mean')  # Moyenne par groupe
```

3. TRANSFORMATION PAR GROUPE (*transform*) :

   - Applique une fonction à chaque groupe, mais conserve la forme originale
   - Ex: moyenne par produit → chaque ligne reçoit la moyenne de son produit
   - Différent de groupby().agg() qui réduit le DataFrame

Pivotage

```python

df.pivot(index='date', columns='produit', values='ventes')
```

4. TABLEAU PIVOT (*pivot*) :

   - Transforme les données de format long à format large
   - Les valeurs deviennent des colonnes
   - Ex: les produits deviennent des colonnes, les dates des lignes
   - Utile pour les analyses croisées et les visualisations

_____________

**Exemple d'application :**  

OPÉRATIONS AVANCÉES : MOBILES, CUMULATIVES ET TRANSFORMATIONS

```python

# Création d'un DataFrame simple
df = pd.DataFrame({
    'date': pd.date_range('2026-01-01', periods=10, freq='D'),
    'produit': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'ventes': [10, 15, 12, 18, 14, 20, 16, 22, 18, 24],
    'prix': [100, 150, 105, 155, 110, 160, 115, 165, 120, 170]
})

print("DATAFRAME ORIGINAL :")
print(df)
```

```python
# 1. Moyenne mobile sur 3 périodes
print("1. MOYENNE MOBILE (3 jours) :")
df['ventes_mm3'] = df['ventes'].rolling(window=3).mean()
print(df[['date', 'ventes', 'ventes_mm3']])


# Résultat : pour chaque ligne, moyenne des 3 dernières ventes
# Lignes 1-2 : NaN (pas assez de données)
# Ligne 3 : moyenne(10, 15, 12) = 12.33
```

```python

# 2. Somme cumulative
print("\n2. SOMME CUMULATIVE :")
df['ventes_cumul'] = df['ventes'].expanding().sum()
print(df[['date', 'ventes', 'ventes_cumul']])

# Ligne 1 : 10
# Ligne 2 : 10 + 15 = 25
# Ligne 3 : 10 + 15 + 12 = 37
```

```python
# 3. Transformation par groupe (conservation des lignes)
print("\n3. MOYENNE PAR PRODUIT :")
df['ventes_moy_produit'] = df.groupby('produit')['ventes'].transform('mean')
print(df[['produit', 'ventes', 'ventes_moy_produit']])

# Toutes les lignes produit A reçoivent la même moyenne
# Toutes les lignes produit B reçoivent la même moyenne
```

```python

# 4. Pivot table (Pivot :  produit → colonnes, date → lignes)
print("\n4. TABLEAU PIVOT :")
df_pivot = df.pivot(index='date', columns='produit', values='ventes')
print(df_pivot)

# Format idéal pour : comparer produits, calculer corrélations
```

```python

# 5. Opérations combinées
print("\n5. COMBINAISONS :")
df['mm3_par_produit'] = df.groupby('produit')['ventes'].rolling(window=3).mean().reset_index(level=0, drop=True)
print(df[['produit', 'ventes', 'mm3_par_produit']])
```

Technique pour enchaîner plusieurs opérations pandas en une seule ligne

**Exemple d'application :**

```python
# Création d'un DataFrame simple
df = pd.DataFrame({
    'nom': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'ville': ['Paris', 'Lyon', 'Paris', 'Marseille', 'Lyon', 'Paris'],
    'ventes': [100, 150, np.nan, 200, -50, 300],
    'commission': [0.1, 0.15, 0.12, 0.18, 0.08, 0.2]
})

print("DATAFRAME ORIGINAL :")
print(df)
```

```python
# Châinage de méthodes
resultat = (df
           .dropna()                     # 1. Supprimer les lignes avec NaN
           .query('ventes > 0')          # 2. Filtrer ventes positives
           .groupby('ville')             # 3. Grouper par ville
           .agg({'commission': 'mean',   # 4. Agréger (moyenne commission)
                 'ventes': 'sum'})       #    et somme des ventes
           .sort_values('ventes',        # 5. Trier par ventes
                       ascending=False)
           .round(2))                    # 6. Arrondir à 2 décimales

print("RÉSULTAT FINAL (chaînage) :")
print(resultat)
```

```python
# Équivalent sans chaînage (plus verbeux) :

print("MÊME TRAITEMENT SANS CHAÎNAGE :")

# Étape par étape
df1 = df.dropna()                       # Supprimer NaN
df2 = df1[df1['ventes'] > 0]            # Filtrer positif
df3 = df2.groupby('ville')              # Grouper
df4 = df3.agg({'commission': 'mean',    # Agréger
               'ventes': 'sum'})
df5 = df4.sort_values('ventes',         # Trier
                      ascending=False)
resultat_sans = df5.round(2)            # Arrondir

print(resultat_sans)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Création d'un DataFrame avec données d'échelles différentes
df = pd.DataFrame({
    'age': [25, 30, 35, 40, 45, 50],          # Échelle : ~30-50
    'salaire': [30000, 40000, 50000, 60000, 70000, 80000],  # Échelle : 30k-80k
    'experience': [2, 5, 8, 12, 15, 20]       # Échelle : 2-20
})

print("DONNÉES ORIGINALES (échelles différentes) :")
print(df)
print(f"\nMoyennes :\n{df.mean()}")
print(f"\nÉcart-types :\n{df.std()}")
```

```python
# Normalisation avec StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("DONNÉES NORMALISÉES (StandardScaler) :")
print(df_scaled)
print(f"\nMoyennes (≈0) :\n{df_scaled.mean().round(3)}")
print(f"\nÉcart-types (=1) :\n{df_scaled.std().round(3)}")
```
