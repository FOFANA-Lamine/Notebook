## 🎲 Partie 1 : Module random standard (Python natif)

📌 Description

Le module random est intégré à Python.
Il permet de générer des nombres pseudo-aléatoires facilement.

✔️ Cas d’usage

- Jeux simples
- Simulations basiques
- Tirages aléatoires

⚠️ Limite

-  Non sécurisé → à éviter pour mots de passe ou tokens



### 1. Fonctions de base pour nombres aléatoires

```python
import random

# 1. random() - Flottant [0.0, 1.0)
x = random.random()  
print("Retourne un flottant entre 0.0 (inclus) et 1.0 (exclus) :",  x)

# 2. uniform(a, b) - Flottant entre a et b (bornes incluses)
x = random.uniform(1.5, 10.5)
print(x)

# 3. randint(a, b) - Entier entre a et b (inclus)
n = random.randint(1, 10)  
print(n)

# 4. randrange([start], stop, [step]) - Comme range() mais aléatoire
a = random.randrange(10)        # 0-9
b = random.randrange(1, 10)     # 1-9
c = random.randrange(0, 100, 5) # 0, 5, 10, ..., 95

print(a)
print(b)
print(c)
```

### 2. Sélections aléatoires dans des séquences

random.choice(seq)

```python
liste = ['a', 'b', 'c', 'd']
element = random.choice(liste)  # Ex: 'c'
print(element)

chaine = "Bonjour"
lettre = random.choice(chaine)  # Ex: 'o'
print(lettre)
```

random.choices(population, weights=None, k=1)

```python
# Avec remise (même élément peut être choisi plusieurs fois)
resultat = random.choices(['a', 'b', 'c'], k=3)  # Ex: ['c', 'c', 'a']
print(resultat)

# Avec pondération
resultat = random.choices(['a', 'b', 'c'], weights=[10, 1, 1], k=4)
# 'a' a 10x plus de chances d'être choisi

print(resultat)
```

random.sample(population, k)

```python
# Sans remise (éléments uniques)
resultat = random.sample(range(100), 5)
# Ex: [42, 17, 89, 23, 56] (tous différents)

print(resultat)
```

### 3. Mélange et réorganisation

random.shuffle(x)

```python
liste = [1, 2, 3, 4, 5]
random.shuffle(liste)  # Modifie la liste sur place
print(liste)  # Ex: [3, 1, 5, 2, 4]


print()
# Pour obtenir une nouvelle liste mélangée sans modifier l'originale
liste_originale = [1, 2, 3, 4]
liste_melangee = random.sample(liste_originale, len(liste_originale))
print(liste_originale, liste_melangee)
```

### 4. Distributions statistiques

random.normalvariate(mu, sigma) ou gauss(mu, sigma)

```python
# Distribution normale (gaussienne)
# mu = moyenne, sigma = écart-type

x = random.normalvariate(0, 1)  
print(x)
```

random.expovariate(lambd)

```python
# Distribution exponentielle

x = random.expovariate(1.5)  # lambda = 1.5
print(x)
```

### Autres distributions disponibles:

betavariate(alpha, beta) - Distribution Beta

triangular(low, high, mode) - Distribution triangulaire

weibullvariate(alpha, beta) - Distribution Weibull

etc.

```python
# 1. gauss(mu, sigma) / normalvariate(mu, sigma) - Distribution normale
x = random.gauss(0, 1)           # -0.234
x = random.normalvariate(0, 1)   # Même chose
print(f"{x}\n")

# 2. expovariate(lambd) - Distribution exponentielle
x = random.expovariate(1.5)
print(f"{x}\n")

# 3. betavariate(alpha, beta) - Distribution Beta
x = random.betavariate(2, 5)
print(f"{x}\n")

# 4. triangular(low, high, mode) - Distribution triangulaire
x = random.triangular(0, 10, 5)
print(f"{x}\n")

# 5. weibullvariate(alpha, beta) - Distribution Weibull
x = random.weibullvariate(1, 1.5)
print(f"{x}\n")


# 6. paretovariate(alpha) - Distribution Pareto
x = random.paretovariate(1)
print(f"{x}\n")

# 7. lognormvariate(mu, sigma) - Distribution log-normale
x = random.lognormvariate(0, 1)
print(f"{x}\n")

# 8. vonmisesvariate(mu, kappa) - Distribution von Mises
x = random.vonmisesvariate(0, 4)
print(f"{x}\n")
```

## 📊 Partie 2 : numpy.random (pour les tableaux)

📌 Description

Le module numpy.random appartient à NumPy.
Il est optimisé pour manipuler des grandes quantités de données.

✔️ Cas d’usage

- Data science
- Machine learning
- Simulation massive

💡 Avantages
- Très rapide
- Génère des tableaux directement
- Nombreuses distributions statistiques

```python
import numpy as np

# Génération de tableaux aléatoires

# 1. rand(d0, d1, ..., dn) - Tableau de flottants [0, 1)
arr1 = np.random.rand(3, 3)  # Matrice 3x3
print(arr1)

print()
# 2. randn(d0, d1, ..., dn) - Distribution normale standard (moyenne 0, écart-type 1)
arr2 = np.random.randn(2, 4)  
print(arr2)

print()
# 3. randint(low, high=None, size=None) - Entiers aléatoires
arr3 = np.random.randint(0, 10, size=(3, 3))  # Matrice 3x3 d'entiers 0-9
print(arr3)


print()
# 4. random(size=None) - Similaire à rand() mais avec size
arr4 = np.random.random((2, 3))  # Matrice 2x3
print(arr4)


print()
# 5. uniform(low=0.0, high=1.0, size=None) - Distribution uniforme
arr5 = np.random.uniform(-1, 1, size=10)
print(arr5)
```

```python

# 6. normal(loc=0.0, scale=1.0, size=None) - Distribution normale
arr6 = np.random.normal(10, 2, size=100)  # Moyenne=10, écart-type=2
print(arr6)
```

```python

# 7. choice(a, size=None, replace=True, p=None) - Choix aléatoire
arr7_1 = np.random.choice([1, 2, 3, 4], size=10)  # 10 choix
arr7_2 = np.random.choice(5, size=3, replace=False)  # Sans remise
print(arr7_1)
print(arr7_2)


print()
# 8. shuffle(x) / permutation(x) - Mélange
arr8 = np.array([1, 2, 3, 4])
np.random.shuffle(arr8)  # Sur place
perm = np.random.permutation(arr8)  # Retourne nouvelle array
print(perm)


# 9. binomial(n, p, size=None) - Distribution binomiale
arr9 = np.random.binomial(10, 0.5, size=100)  # 100 tirages Binomial(10, 0.5)
print(arr9)
```

```python

# 10. poisson(lam=1.0, size=None) - Distribution de Poisson
arr10 = np.random.poisson(5, size=50)  # Poisson(λ=5)
print(arr10)


print()
# 11. exponential(scale=1.0, size=None) - Distribution exponentielle
arr11 = np.random.exponential(2, size=100)  # β=2
print(arr11)
```

```python
# 12. beta(a, b, size=None) - Distribution Beta
arr12 = np.random.beta(2, 5, size=50)
print(arr12)


print()
# 13. gamma(shape, scale=1.0, size=None) - Distribution Gamma
arr13 = np.random.gamma(2, 2, size=50)
print(arr13)
```

### Nouvelle API NumPy (recommandée depuis 1.17)

```python
# Créer un générateur
rng = np.random.default_rng(seed=42)

# Utiliser les méthodes
arr1 = rng.random((3, 3))      # Au lieu de np.random.rand()
print(arr1)

print()
arr2 = rng.standard_normal((3, 3))  # Au lieu de np.random.randn()
print(arr2)

print()
arr3 = rng.integers(0, 10, size=(3, 3))
print(arr3)

print()
arr4 = rng.normal(0, 1, size=50)
print(arr4)
```

## 🔐 Partie 3 : Module secrets (cryptographie)

📌 Description

Le module secrets est conçu pour la sécurité.
Il génère des nombres aléatoires cryptographiquement sûrs.
il résiste aux attaques et convient aux mots de passe, jetons d'authentification, clés de chiffrement, etc.

✔️ Cas d’usage

- Mots de passe
- Tokens d’authentification
- Clés de sécurité

🔒 Avantage clé

- Résistant aux attaques (contrairement à random)



### 1. secrets.token_bytes(nbytes=None)

Génère des octets aléatoires sécurisés

```python
import secrets

# Générer 16 octets (128 bits) pour une clé de chiffrement
key = secrets.token_bytes(16)
print(key)    # b'\x7f\xb4...' (bytes bruts)

# Usage typique : clé AES-128
from cryptography.fernet import Fernet
key = secrets.token_bytes(32)  # 256 bits pour AES-256
print(key)
```

Pourquoi l'utiliser ?
Pour créer des clés de chiffrement ou des données binaires sensibles.

### 2. secrets.token_hex(nbytes=None)

Génère une chaîne hexadécimale sécurisée

```python
# Générer 16 octets → 32 caractères hexadécimaux
token = secrets.token_hex(16)
print(token)  

# Cas d'usage : clé API, lien de réinitialisation de mot de passe
reset_link = f"https://monsite.com/reset?token={token}"
```

Caractéristiques :

- Chaque octet donne 2 caractères hexadécimaux

- Taille prévisible (2 × nbytes)

- Facile à copier/coller ou stocker en base de données



### 3. secrets.token_urlsafe(nbytes=None)

Génère un token encodé en Base64 (compatible URL)

```python
# Générer un token sans caractères problématiques (+/=)
token = secrets.token_urlsafe(16)
print(token)  

# Usage réel : jeton CSRF, session utilisateur
csrf_token = secrets.token_urlsafe(32)
print(csrf_token)
```

Avantage :

- Utilise l'alphabet A-Z, a-z, 0-9, - et _

- Pas de /, +, = qui doivent être encodés en URL

- Idéal pour les paramètres HTTP ou cookies

### 4. secrets.choice(sequence)

Choix aléatoire cryptographique

```python
# Choisir un mot de passe simple parmi une liste
base = ['chien', 'chat', 'soleil', 'pluie']
mot = secrets.choice(base) + str(secrets.randbelow(100))
print(mot)  # "soleil42"


# Générer un caractère aléatoire
import string
alphabet = string.ascii_letters + string.digits
char = secrets.choice(alphabet)
print(char)
```

Alternative à random.choice :

random.choice est **prévisible** → à ne **JAMAIS** utiliser pour la sécurité

### 5. secrets.randbelow(n)

Entier aléatoire entre 0 (inclus) et n (exclu)

```python
# Nombre entre 0 et 99
num = secrets.randbelow(100)

# Générer un code à 6 chiffres (entre 100000 et 999999)
code = secrets.randbelow(900000) + 100000
print(code)


# Mot de passe avec chiffres
digits = ''.join(str(secrets.randbelow(10)) for _ in range(6))
print(digits)  # "483729"
```

Différence avec *random.randint* :

*secrets.randbelow* est **imprévisible** même si l'attaquant connaît la graine

## Exemple concret : générateur de mot de passe robuste

```python
import secrets
import string

def generer_mdp(longueur=12):
    caracteres = string.ascii_letters + string.digits + "!@#$%^&*"
    return ''.join(secrets.choice(caracteres) for _ in range(longueur))

# Générer 3 mots de passe forts
for _ in range(3):
    print(generer_mdp(16))
```

```python
import time
import string    #  pour générer l'alphabet 
import statistics    #  pour calculer la médiane des temps


SECRET = "Lamté"
     

# Fonction vulnérable (comparaison non sécurisée)
def insecure_compare(user_input):
    # Simulation d'une fuite temporelle de 2ms par caractère correct
    for i in range(min(len(user_input), len(SECRET))):
        if user_input[i] != SECRET[i]:
            return False
        time.sleep(0.002)  # Fuite volontaire de 2ms
    return len(user_input) == len(SECRET)


""" Explication :

Cette fonction compare l'entrée utilisateur avec un secret.
user_input : C'est la chaîne de caractères que l'attaquant propose pour tester.

____________
for i in range(min(len(user_input), len(SECRET))):

On boucle sur le plus petit des deux pour éviter de comparer au-delà de la longueur la plus courte.
Exemple concret :
SECRET = "ABC"  # longueur 3
user_input = "AB"  # longueur 2
min(2, 3) = 2  # On compare seulement les 2 premiers caractères
_______________

if user_input[i] != SECRET[i]:
            return False

            
- Compare caractère par caractère 
- Dès qu'un caractère est différent -> retourne False immédiatement
- Aucun délai pour ce cas (retour instantané)

Exemple : text

user_input = "AXC"
SECRET = "ABC"

i=0: 'A' vs 'A' → OK, continue
i=1: 'X' vs 'B' → DIFFÉRENT -> return False (instantané)

____________________________

 time.sleep(0.002)  # Fuite volontaire de 2ms

- time.sleep(0.002) : Pause de 2 millisecondes (0.002 secondes)
- C'est la vulnérabilité !
- On ne fait cette pause que si le caractère est correct
- Plus on a de caractères corrects, plus la fonction prend du temps

Visualisation : 

SECRET = "ABC"

"ABC" → ✅ 'A' (pause 2ms) → ✅ 'B' (pause 2ms) → ✅ 'C' (pause 2ms) → 6ms

"ABX" → ✅ 'A' (pause 2ms) → ✅ 'B' (pause 2ms) → ❌ 'X' ≠ 'C' → 4ms

"AXC" → ✅ 'A' (pause 2ms) → ❌ 'X' ≠ 'B' → 2ms

"XBC" → ❌ 'X' ≠ 'A' → 0ms

________________________________

return len(user_input) == len(SECRET)

- Si on a parcouru toute la boucle sans trouver d'erreur
- Alors on vérifie que les deux chaînes ont la même longueur
- Retourne True seulement si les deux sont identiques ET de même longueur


Exemples :
SECRET = "ABC"
user_input = "ABC"  → len(3) == len(3) → True ✅
user_input = "AB"   → len(2) == len(3) → False ❌ (mais on a déjà vérifié les 2 premiers)
user_input = "ABCD" → len(4) == len(3) → False ❌

______________________________

Le point clé à comprendre
L'attaquant peut mesurer le temps et en déduire combien de caractères sont corrects :

Temps ~0ms   → 0 caractère correct   → 1ère lettre fausse
Temps ~2ms   → 1 caractère correct   → 1ère lettre bonne, 2ème fausse
Temps ~4ms   → 2 caractères corrects → 1ère et 2ème bonnes, 3ème fausse
Temps ~6ms   → 3 caractères corrects → MOT DE PASSE COMPLET ! 🎯


________________________________

Que représente user_input ?
C'est la tentative de l'attaquant à chaque étape :

# Étape 1 : Deviner la 1ère lettre
insecure_compare("A")   # temps ~2ms → bonne !
insecure_compare("B")   # temps ~0ms → mauvaise
insecure_compare("C")   # temps ~0ms → mauvaise

# Étape 2 : Deviner la 2ème lettre (sachant que 1ère = 'A')
insecure_compare("AA")  # temps ~2ms → 2ème fausse
insecure_compare("AB")  # temps ~4ms → 2ème bonne ! (2 caractères corrects)
insecure_compare("AC")  # temps ~2ms → 2ème fausse

# Étape 3 : Deviner la 3ème lettre (sachant que début = 'AB')
insecure_compare("ABA") # temps ~4ms → 3ème fausse
insecure_compare("ABB") # temps ~4ms → 3ème fausse
insecure_compare("ABC") # temps ~6ms → 3ème bonne ! → mot de passe trouvé



C'est ainsi que l'attaquant reconstruit le mot de passe caractère par caractère !

"""




# --- Paramètres de l'attaque ---

# ALPHABET = string.ascii_uppercase + string.ascii_lowercase + string.digits 
ALPHABET = (
    string.ascii_uppercase +
    string.ascii_lowercase +
    string.digits +
    string.punctuation +
    "éèêëàâäùûüîïôöç" 
    # + "".join(chr(i) for i in range(0x00C0, 0x017F))   Plus complète(Unicode latin)
)

NB_ESSAIS = 15  # Nombre de mesures par caractère (plus c'est haut, plus c'est fiable)

def mesurer_temps_robuste(guess):
    mesures = []
    for _ in range(NB_ESSAIS):
        start = time.perf_counter()
        insecure_compare(guess)
        end = time.perf_counter()
        mesures.append(end - start)
        
    # On utilise la médiane pour ignorer les pics de lag du processeur
    return statistics.median(mesures)




""" Explication :

Role général : Cette fonction mesure plusieurs fois le temps de comparaison d'une tentative (guess) pour obtenir 
une mesure fiable et robuste au bruit du système.

guess : C'est la tentative que l'attaquant veut tester.
Exemple : "A", "AB", "ABC", etc.
guess = ce qu'on pense être le début du mot de passe


time.perf_counter() : Horloge de haute précision (microsecondes)

_________

insecure_compare(guess)

Exécute la fonction vulnérable avec la tentative guess
C'est là que le temps s'écoule (selon le nombre de caractères corrects)

__________

Exemple concret :
NB_ESSAIS = 3
guess = "AB"

Mesure 1 : start=100.000, end=100.004 → 0.004s (4ms) → mesures = [0.004]
Mesure 2 : start=100.010, end=100.014 → 0.004s → mesures = [0.004, 0.004]
Mesure 3 : start=100.020, end=100.025 → 0.005s (coup de processeur) → mesures = [0.004, 0.004, 0.005]

_______________

 return statistics.median(mesures)

- statistics.median() : Calcule la médiane des mesures
- La médiane = valeur du milieu quand on trie les temps

Pourquoi la médiane ?

Mesures(triées)	           Moyenne	                Médiane
[0.004, 0.004, 0.005]	   0.00433	                0.004

La médiane ignore les pics aberrants (quand le processeur est occupé ailleurs)


Exemple complet d'exécution :

SECRET = "ABC"
NB_ESSAIS = 3

# Test de "A"
print(mesurer_temps_robuste("A"))
# Mesure 1: 2ms
# Mesure 2: 2ms
# Mesure 3: 2ms
# → médiane = 2ms

# Test de "AB"  
print(mesurer_temps_robuste("AB"))
# Mesure 1: 4ms
# Mesure 2: 4ms
# Mesure 3: 15ms (coup de processeur)
# → médiane = 4ms (ignore le pic)

# Test de "ABC"
print(mesurer_temps_robuste("ABC"))
# Mesure 1: 6ms
# Mesure 2: 6ms
# Mesure 3: 6ms
# → médiane = 6ms



L'attaquant utilise cette fonction pour obtenir un temps fiable pour chaque guess, 
puis compare les temps pour deviner le bon caractère.

"""



def attaque_recursive(prefix=""):
    """
    Algorithme récursif avec backtracking pour garantir 95%+ de succès.
    """
    if len(prefix) == len(SECRET):
        # Vérification finale
        if insecure_compare(prefix):
            return prefix
        return None

    print(f"🔍 Recherche position {len(prefix)+1} (préfixe actuel: '{prefix}')")
    
    scores = []
    for c in ALPHABET:
        test = prefix + c
        t = mesurer_temps_robuste(test)
        scores.append((t, c))
    
    # On trie : les temps les plus longs en premier
    scores.sort(key=lambda x: x[0], reverse=True)

    # On teste les 3 meilleurs candidats (au cas où le 1er soit une erreur due au bruit)
    for i in range(3):
        candidat_char = scores[i][1]
        resultat = attaque_recursive(prefix + candidat_char)
        if resultat:
            return resultat
            
    return None




"""  Rôle général
Cette fonction reconstruit le mot de passe caractère par caractère en utilisant la récursivité et le backtracking.
Elle teste chaque caractère possible, mesure le temps, et continue avec les meilleurs candidats.

__________
def attaque_recursive(prefix=""):

prefix : La partie du mot de passe déjà découverte
Commence vide "", puis "A", puis "AB", etc.
C'est comme une accumulation des lettres trouvées
_____________

if len(prefix) == len(SECRET):
Condition d'arrêt : Quand le préfixe a la même longueur que le secret
Exemple : SECRET = "ABC" → on s'arrête quand len(prefix) = 3

_________________

if insecure_compare(prefix):
    return prefix
return None

Vérifie si prefix est le bon mot de passe

Si oui → retourne le mot de passe trouvé
Si non → retourne None (chemin invalide)


______________

scores = []
for c in ALPHABET:
    test = prefix + c
    t = mesurer_temps_robuste(test)
    scores.append((t, c))

    
Phase de test : On essaie TOUS les caractères de l'alphabet
Pour chaque caractère c, on construit test = prefix + c
On mesure le temps de réponse avec mesurer_temps_robuste(test)
On stocke le couple (temps, caractère) dans scores


Exemple avec prefix = "A" :
Text

ALPHABET = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

Test "AA" → temps 2ms → scores.append((2, 'A'))
Test "AB" → temps 4ms → scores.append((4, 'B'))  ← plus long !
Test "AC" → temps 2ms → scores.append((2, 'C'))
...
Test "AZ" → temps 2ms → scores.append((2, 'Z'))
    
_____________
scores.sort(key=lambda x: x[0], reverse=True)

- Tri décroissant : Les temps les plus longs en premier
- lambda x: x[0] → on trie selon le premier élément du tuple (le temps)
- reverse=True → du plus grand au plus petit

Exemple de tri :
text
Avant tri : [(2,'A'), (4,'B'), (2,'C'), (1,'D')]
Après tri : [(4,'B'), (2,'A'), (2,'C'), (1,'D')]
→ 'B' est le meilleur candidat (temps le plus long)

_______________

for i in range(3):
    candidat_char = scores[i][1]
    resultat = attaque_recursive(prefix + candidat_char)
    if resultat:
        return resultat

Backtracking : On teste les 3 meilleurs candidats (et pas seulement le 1er)
Pourquoi 3 ? Pour gérer le bruit de mesure (parfois le 1er est faux à cause d'un pic CPU)
Pour chaque candidat, on appelle récursivement la fonction avec le nouveau préfixe
Si un chemin réussit (resultat n'est pas None), on le retourne


Arbre de recherche :

                    prefix = ""
                       |
            ┌──────────┼──────────┐
            |          |          |
         candidat1   candidat2   candidat3
            |          |          |
          "A"         "B"         "C"
            |          |          |
         ┌──┼──┐    ┌──┼──┐    ┌──┼──┐
         A  B  C    A  B  C    A  B  C
         |  |  |    |  |  |    |  |  |
        "AA" ...  "BA" ...  "CA" ...

____________

return None

Si aucun des 3 meilleurs candidats ne mène au mot de passe
Retourne None pour indiquer que ce chemin est un cul-de-sac


"""



# --- Exécution ---
print("="*50)
print("🔓 DÉBUT DE L'ATTAQUE TEMPORELLE HAUTE PRÉCISION")
print("="*50)

debut_total = time.time()
mot_de_passe_final = attaque_recursive()
fin_total = time.time()

if mot_de_passe_final:
    print("\n" + "="*50)
    print(f"✅ SUCCÈS : '{mot_de_passe_final}'")
    print(f"⏱️ Temps total : {fin_total - debut_total:.2f} secondes")
    print("="*50)
else:
    print("\n❌ Échec de la reconstruction.")
```

### Pourquoi cette version est-elle "puissante" ?

Utilisation de *statistics.median* : La moyenne est polluée si ton antivirus ou Windows décide de travailler pendant une microseconde. La médiane, elle, reste stable car elle ignore les valeurs extrêmes.

*Backtracking (Récursion)* : C'est le secret des 95 %. Si le script se trompe au 3ème caractère à cause d'un ralentissement du PC, il ne va pas rester bloqué. Il va tester le "2ème meilleur choix" de la position précédente.

*Filtrage du Bruit* : Avec *NB_ESSAIS = 15*, on réduit statistiquement la probabilité d'erreur à un niveau quasi nul sur un délai de 2ms.

**Note pédagogique** : Dans un vrai système, la fuite est de l'ordre de la nanoseconde, pas de la milliseconde. Il faudrait alors des millions d'essais pour extraire le mot de passe, mais la logique mathématique resterait exactement la même !
