## 2. Symboles et Expressions de Base

```python
import sympy as sp
from sympy import *    #  symbols, Function, diff, Eq, solve , I, re, im, abs, arg, conjugate, polar, ....

# Déclarer des symboles
x, y, z, k = symbols('x y z k')
a, b, c = symbols('a b c', real=True)  # Nombres réels
n = symbols('n', integer=True)         # Entier
t = symbols('t', positive=True)        # Positif

# Expressions symboliques
expr1 = x**2 + 2*x + 1     
expr2 = sin(x) + cos(y)
expr3 = exp(x) * log(y)

# Affichage LaTeX
print(sp.latex(expr1))  # x^{2} + 2 x + 1
print(sp.latex(expr2))  #  \sin{\left(x \right)} + \cos{\left(y \right)}
print(sp.latex(expr3))  #  e^{x} \log{\left(y \right)}

# Affichage Simple
print(expr1)
print(expr2)
print(expr3)



# Substitution
expr = x**2 + y
result = expr.subs({x: 2, y: 3})  # 7
print(result)

result2 = expr.subs(x, y)         
print(result2)   # y² + y
```

## 3. Simplification et Manipulation

```python
# Simplification 
expr = (x**2 - 1)/(x - 1)
simplified = simplify(expr)  
print(simplified)   # x + 1

# Expansion (Développement) )
expr = (x + 1)**3
expanded = expand(expr)  # x³ + 3x² + 3x + 1
print(expanded)

# Factorisation
expr = x**2 - 4
factored = factor(expr)  
print(factored)   # (x - 2)(x + 2)

# Collecte de termes
expr = x*y + x - 3 + 2*x**2 - z*x**2 + x**3
collected = collect(expr, x)    # Regroupé par puissances de x
print(collected)          # x**3 + x**2*(2 - z) + x*(y + 1) - 3

# Fonctions trigonométriques
expr = sin(x)**2 + cos(x)**2
trig_simp = trigsimp(expr) 
print(trig_simp)           # 1

# Fractions partielles (Décomposition en éléments simple )
expr = 1/((x+2)*(x+1))
apart_expr = apart(expr)  
print(apart_expr)          # -1/(x+2) + 1/(x+1)
```

## 4. Résolution d'Équations

```python
# Équations algébriques
solutions = solve(x**2 - 4, x)           
print(solutions)  # [-2, 2]

solutions = solve([x + y - 3, x - y - 1], [x, y])    # , dict=True
print(solutions)   # {x: 2, y: 1}


# Équations différentielles
f = Function('f')
eq = Eq( f(x).diff(x, x) - f(x), exp(x) )  
solution = dsolve(eq)  # Solution générale
print(solution)


# Systèmes linéaires
A = Matrix([[1, 2], [3, 4]])
b = Matrix([5, 6])
solution = A.LUsolve(b)  # Solution du système Ax = b
print(solution)
```

## 5. Calcul Différentiel

```python
# Dérivées
f = x**3 + 2*x**2 + x
f_prime = diff(f, x)             # 3x² + 4x + 1
f_double = diff(f, x, 2)         # 6x + 4
partial = diff(x**2 * y, x, y)   # Dérivée partielle mixte


# Développement de Taylor
a = series(exp(x), x, 0, 5)         
print(a)      # 1 + x + x²/2 + x³/6 + x⁴/24 + O(x⁵)

# Limites
b = limit(sin(x)/x, x, 0)           
print(b)         # 1

c = limit((1 + 1/x)**x, x, oo)      
print(c)                   # e (infini)

```

## 6. Calcul Intégral

```python
# Intégrales indéfinies
integrate(x**2, x)               # x³/3
integrate(sin(x), x)             # -cos(x)
integrate(1/x, x)                # log(x)

# Intégrales définies
integrate(x**2, (x, 0, 1))       # 1/3
integrate(exp(-x**2), (x, -oo, oo))  # √π

# Intégrales multiples
integrate(x*y, (x, 0, 1), (y, 0, 1))  # 1/4


# Intégrales impropres
integrate(1/x**2, (x, 1, oo))    # 1
```

## 7. Algèbre Linéaire

```python
# Matrices
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[5, 6], [7, 8]])

# Opérations
A + B           # Addition
A * B           # Multiplication
A**2            # Carré
A.inv()         # Inverse
A.det()         # Déterminant
A.T             # Transposée

# Valeurs propres et vecteurs propres
A.eigenvals()   # Valeurs propres
A.eigenvects()  # Vecteurs propres


# Forme échelonnée
A.rref()        # Forme échelonnée réduite
```

## 8. Nombres Complexes

```python
z = 3 + 4*I
re(z)           # Partie réelle: 3
im(z)           # Partie imaginaire: 4
abs(z)          # Module: 5
arg(z)          # Argument: atan(4/3)
conjugate(z)    # Conjugué: 3 - 4i

```

## 9. Série et Sommes

```python
# Sommes finies
summation(k, (k, 1, n))          # n(n+1)/2
summation(1/2**k, (k, 0, oo))    # 2

# Produits
product(k, (k, 1, n))            # n! (factorielle)


# Séries
series(cos(x), x, 0, 6)          # 1 - x²/2 + x⁴/24 + O(x⁶)
```

## 10. Statistiques et Probabilités

```python
from sympy.stats import *

# Variables aléatoires
X = Normal('X', 0, 1)      # Normale N(0,1)
Y = Exponential('Y', 1)    # Exponentielle λ=1
Z = Poisson('Z', 3)        # Poisson λ=3

# Calculs de probabilité
P(X > 0)                   # 1/2
E(X**2)                    # Espérance: 1
variance(X)                # Variance: 1
density(X)(0)              # Densité en 0: 1/√(2π)

```

# 13. Exemples Complets

```python

# 1. Résolution physique : Mouvement harmonique
t, ω, A, φ = symbols('t ω A φ', real=True)
x = A*cos(ω*t + φ)
v = diff(x, t)          # Vitesse
a = diff(v, t)          # Accélération

# Vérifier l'équation différentielle
eq_check = simplify(a + ω**2 * x)   # Doit être 0
print(eq_check)   
```

```python
# 2. Optimisation : Trouver un extremum
x = symbols('x')
f = x**3 - 6*x**2 + 9*x + 1
f_prime = diff(f, x)

critical_points = solve(f_prime, x)  # Points critiques
f_double = diff(f, x, 2)

# Évaluer la nature des points critiques
for point in critical_points:
    second_deriv = f_double.subs(x, point)
    print(f"Point {point}: f'' = {second_deriv}")
```

```python

# 3. Calcul de volume
# METHODE 1 : 


r, h = symbols('r h', positive=True)
V = pi * r**2 * h                    # Volume du cylindre
surface = 2*pi*r*h + 2*pi*r**2       # Surface totale

# Contrainte : V = 100
sol_h = solve(Eq(V, 100), h)[0]      # Exprimer h en fonction de r
print(f"h en fonction de r: {sol_h}")

# Fonction de surface en fonction de r seul
S_r = simplify(surface.subs(h, sol_h))
print(f"Surface S(r) = {S_r}")

# Trouver le minimum : dériver et annuler la dérivée
dS_dr = diff(S_r, r)
print(f"dS/dr = {dS_dr}")

# Résoudre dS/dr = 0
critical_points = solve(Eq(dS_dr, 0), r)
print(f"Points critiques: {critical_points}")

# Filtrer les points positifs
optimal_r = [cp for cp in critical_points if cp > 0][0]
print(f"Rayon optimal: r = {optimal_r}")

# Calculer h optimal
optimal_h = sol_h.subs(r, optimal_r)
print(f"Hauteur optimale: h = {optimal_h}")

# Surface minimale
min_surface = S_r.subs(r, optimal_r)
print(f"Surface minimale: S = {min_surface}")

# Vérification avec la dérivée seconde
d2S_dr2 = diff(S_r, r, 2)
second_deriv = d2S_dr2.subs(r, optimal_r)
print(f"d²S/dr² au point critique = {second_deriv}")
print(f"Minimum ? {second_deriv > 0}")
```

```python
# METHODE 2 : Approche numérique vec SciPy

import numpy as np
from scipy.optimize import minimize

def surface_function(r):
    # V = πr²h = 100 ⇒ h = 100/(πr²)
    h = 100 / (np.pi * r**2)
    return 2 * np.pi * r * h + 2 * np.pi * r**2

# Optimisation numérique
result = minimize(surface_function, x0=2, bounds=[(0.1, 10)])
print(f"\nOptimisation numérique (SciPy):")
print(f"r optimal ≈ {result.x[0]}")
print(f"h optimal ≈ {100/(np.pi*result.x[0]**2)}")
print(f"Surface minimale ≈ {result.fun}")
```

```python
# Méthode analytique

r, h = symbols('r h', positive=True)
V = pi * r**2 * h
S = 2*pi*r*h + 2*pi*r**2

# 1. Exprimer h depuis V=100
h_expr = solve(Eq(V, 100), h)[0]

# 2. Surface en fonction de r
S_r = simplify(S.subs(h, h_expr))

# 3. Trouver le minimum (dérivée = 0)
dS_dr = diff(S_r, r)
optimal_r_solutions = solve(Eq(dS_dr, 0), r)

# 4. Prendre la solution positive
optimal_r = [sol for sol in optimal_r_solutions if sol > 0][0]
optimal_h = h_expr.subs(r, optimal_r)

print(f"Solution analytique:")
print(f"r = {optimal_r}")
print(f"h = {optimal_h}")
print(f"h/r = {optimal_h/optimal_r}")
print(f"Pour un cylindre, h = 2r donne la surface minimale")
```

## 🎯 Recommandations Finales

### **Utilisez SymPy quand** :

Vous avez besoin de résultats exacts (fractions, racines)

Vous travaillez avec des variables symboliques

Vous faites du calcul différentiel/intégral symbolique

Vous manipulez des expressions algébriques complexes

Vous générez du code LaTeX


### **Utilisez NumPy quand** :
Vous travaillez avec de grandes quantités de données

Vous avez besoin de performance

Vous faites des calculs numériques approximatifs

Vous utilisez des tableaux/matrices

Vous intégrez avec d'autres bibliothèques scientifiques



## 🎯 SymPy en Bref
SymPy est l'outil idéal pour les mathématiques symboliques en Python. Contrairement à NumPy qui excelle dans le calcul numérique, SymPy préserve la forme exacte des expressions mathématiques.
